```python
import matplotlib.pyplot as plt
import numpy as np
import random

f=open("hw1_train.txt",mode='r')
data=f.read()
data=data.split('\n')
n=0
for i in data:
    data[n]=i.split('\t')
    data[n]=[float(x) for x in data[n]]
    data[n]=[(data[n][0],data[n][1],data[n][2],data[n][3],data[n][4],data[n][5],data[n][6],data[n][7],data[n][8],data[n][9]),data[n][10]]
    n+=1
data=np.asarray(data)
#print(data)
f.close()
```

```python
def pla(data,x0):
    w=np.zeros(11)
    count=0
    c=0
    while(c!=500):
        ii=random.randint(0,99)
        xx=np.array(data[ii][0])
        x=np.concatenate(([x0],xx))
        s=data[ii][1]
        y=w.T.dot(x)
        ii+=1
        if y*s<=0:
            w+=s*x
            count+=1
            c=0
        else:
            c+=1
    return(w,count)
```

```python
#probelm16&17
#x0=1
updates=[]
w0=[]
for i in range(1000):
    (w,count)=pla(data,1)
    updates.append(count)
    w0.append(w[0])
print(w)
print(np.median(updates))
print(w0)
```
